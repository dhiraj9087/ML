```python
import numpy as np
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])     #if one of the axis not specified it will get default values as 0,1,2,3....
plt.plot(xpoints,ypoints,marker = "<")    #marker for highlighting each point
plt.show()
```

```python
############# markers ###########
'o'	Circle	
'*'	Star	
'.'	Point	
','	Pixel	
'x'	X	
'X'	X (filled)	
'+'	Plus	
'P'	Plus (filled)	
's'	Square	
'D'	Diamond	
'd'	Diamond (thin)	
'p'	Pentagon	
'H'	Hexagon	
'h'	Hexagon	
'v'	Triangle Down	
'^'	Triangle Up	
'<'	Triangle Left	
'>'	Triangle Right	
'1'	Tri Down	
'2'	Tri Up	
'3'	Tri Left	
'4'	Tri Right	
'|'	Vline	
'_'	Hline
```
