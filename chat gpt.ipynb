```python
# Tic-Tac-Toe game in Python

board = [" "," "," "," "," "," "," "," "," "]

def print_board():
    print(" "+board[0]+" | "+board[1]+" | "+board[2]+" ")
    print("---+---+---")
    print(" "+board[3]+" | "+board[4]+" | "+board[5]+" ")
    print("---+---+---")
    print(" "+board[6]+" | "+board[7]+" | "+board[8]+" ")

def check_winner():
    if(board[0] == board[1] and board[1] == board[2] and board[0] != " "):
        return board[0]
    if(board[3] == board[4] and board[4] == board[5] and board[3] != " "):
        return board[3]
    if(board[6] == board[7] and board[7] == board[8] and board[6] != " "):
        return board[6]
    if(board[0] == board[3] and board[3] == board[6] and board[0] != " "):
        return board[0]
    if(board[1] == board[4] and board[4] == board[7] and board[1] != " "):
        return board[1]
    if(board[2] == board[5] and board[5] == board[8] and board[2] != " "):
        return board[2]
    if(board[0] == board[4] and board[4] == board[8] and board[0] != " "):
        return board[0]
    if(board[2] == board[4] and board[4] == board[6] and board[2] != " "):
        return board[2]
    if " " not in board:
        return "Tie"
    return None

player = "X"
while True:
    print_board()
    print("Player "+player+", make your move (1-9):")
    move = int(input())-1
    if(move>=0 and move<=8 and board[move] == " "):
        board[move] = player
    else:
        print("Invalid move!")
        continue
    winner = check_winner()
    if winner != None:
        print_board()
        print(winner+" wins!")
        break
    if player == "X":
        player = "O"
    else:
        player = "X"
```

```python
import random

# Create an array of roll numbers for students who attended the training program
roll_numbers = [i for i in range(1, 51)]
random.shuffle(roll_numbers)
print("Roll numbers of students who attended the training program:", roll_numbers)

# Define a function for linear search
def linear_search(roll_numbers, search_number):
    for i in range(len(roll_numbers)):
        if roll_numbers[i] == search_number:
            return True
    return False

# Define a function for sentinel search
def sentinel_search(roll_numbers, search_number):
    roll_numbers.append(search_number)
    i = 0
    while roll_numbers[i] != search_number:
        i += 1
    roll_numbers.pop()
    return i != len(roll_numbers)

# Test the search functions
search_number = int(input("Enter a roll number to search for: "))

if linear_search(roll_numbers, search_number):
    print("The student with roll number", search_number, "attended the training program (linear search)")
else:
    print("The student with roll number", search_number, "did not attend the training program (linear search)")

if sentinel_search(roll_numbers, search_number):
    print("The student with roll number", search_number, "attended the training program (sentinel search)")
else:
    print("The student with roll number", search_number, "did not attend the training program (sentinel search)")
```
